```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Signal Parameters.
t_min = 0
t_max = 1
t = np.linspace(t_min, t_max, 1000)
f = 3
```

```python
# The Signal
y_t = np.sin(2 * np.pi * f * t)
plt.plot(t, y_t)
plt.title('A pure Sine Wave Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
```

```python
# Quantization Process

min_value = np.min(y_t)
max_value = np.max(y_t)
signal_range = max_value - min_value

num_of_bits = 2
no_of_levels = 2**num_of_bits
step = signal_range / no_of_levels

quantized_index = np.floor((y_t - min_value) / step)

clipped_index = np.clip(quantized_index, 0, no_of_levels - 1)

# Employing Mid-rise Quantization
quantized_value =  min_value + ((clipped_index + 0.5) * step)

# Final Plot
plt.step(t, quantized_value, 'r', label='Quantized Signal')
plt.plot(t, y_t, label='Original Signal')
plt.ylabel('Quantized Values (Q)')
plt.xlabel('Samples (n)')
plt.title('2-Bit Quantized Signal')
plt.legend()
plt.grid(True)
```

```python
# Excluding the mid-rise quantization

quantized_value =  min_value + (clipped_index * step)

plt.step(t, quantized_value, 'r', label='Quantized Signal')
plt.plot(t, y_t, label='Original Signal')
plt.ylabel('Quantized Values (Q)')
plt.xlabel('Samples (n)')
plt.title('2-Bit Quantized Signal')
plt.legend()
plt.grid(True)
```

This notebook provides, in code, how the quantization process is implemented. The 'Quantization Notes' provides the theoretical background and detailed explanation of the procedure. While the notes uses 8 quantization levels, this notebook uses 4 levels: a 2-bit quantizer.
The np.clip() function is used to map values that exceed the range to the maximum or minimum level. Its basic form is np.clip(array, a_min, a_max)
