```python
import numpy as np
import scipy.signal as sp
import matplotlib.pyplot as plt
```

```python
# Signal parameters

t_min = 0 
t_max = 1 
f = 4 #signal frequency
```

```python
# Sampling at 100 Hz

fs1 = 100
t1 = np.linspace(t_min, t_max, fs1, endpoint=False)

x1 = np.sin(2 * np.pi * f * t1)
plt.stem(t1, x1)
plt.plot(t1, x1)
plt.title('4 Hz sine wave sampled at 100 Hz')
plt.ylabel('Amplitude')
plt.xlabel('Time (s)')
```

```python
# Sampling at 3 Hz

fs2 = 3
t2 = np.linspace(t_min, t_max, fs2, endpoint=False)

x2 = np.sin(2 * np.pi * f * t2)
plt.stem(t2, x2)
plt.plot(t2, x2)
plt.title('4 Hz sine wave sampled at 3 Hz')
plt.ylabel('Amplitude')
plt.xlabel('Time (s)')
```

```python
# Resampling the low-rate signal for smooth plotting
r_x, s_t = sp.resample(x2, 1000, t2)

plt.plot(s_t, r_x, label='Resampled Signal')
plt.plot(t1, x1, label='Original Signal')
plt.ylabel('Amplitude')
plt.xlabel('Time (s)')
plt.title('Comparison of Original and Resampled Signals')
plt.legend()
```

With sampling and the Shannon sampling theorem discussed, the code in this notebook provides graphical support for the explanation.

In the first instance, the signal is sampled at 100 Hz, meaning we are taking 100 samples per second. This proves to be sufficient, since a line joining the points within one second, perfectly traces out a sine curve.

In the second instance, the signal is sampled at a frequency lower than that required by the sampling theorem. For a 4 Hz signal, the minimum required sampling frequency is 8 Hz, but the signal is sampled at 3 Hz. Hence, the reconstructed signal appears to have a frequency of 1 Hz. Aliasing has occured.
