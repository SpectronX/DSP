```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defining the parameters

t_min = 0 #minimum time
t_max = 1 #maximum time
f = 1 #frequency of the signal
fs = 50 #sampling frequency
Ts = 1 / fs #sampling period
t = np.arange(t_min, t_max, Ts) #time vector from 0 to 1 with step size Ts
```

```python
# Signal
x_t = np.sin(2 * np.pi * f * t)

# Plotting the signal
plt.stem(t, x_t)
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.title("Plot of the signal")
```

Numpy is a python library that provides support for numerical and mathematical operations. Matplotlib is also a library used for creating visualizations: graphs and plots.

In defining the parameters, f the frequency of the signal, determines the number of oscillations in one second. In this case, 1. 

fs is the sampling frequency. It's asking, how many samples/discrete-time values of the signal should be captured within one second? In our case, 50. Therefore, at 50 discrete points, the magnitude of the signal was captured.

Ts is the sampling period. It's giving us the time interval for which the discrete values were captured. Here, it's every 0.02 seconds.
