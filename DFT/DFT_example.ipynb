### The Discrete Fourier Transform (DFT)

- The DFT is one among the members of the Fourier transform family. 

- It assumes an N sample sequence signal as one period of an aperiodic signal and treating the signal as periodic. The DFT deals with only finite-length sequences, while the Discrete-time fourier transform (DTFT) handles aperiodic discrete signals of infinite duration. The DFT simply samples the DTFT.

- The DFT can be calculated using the formula:

- X[k] =  $\sum_{n = 0} ^ {N - 1} x[n] e^{-i 2 \pi kn/N}$

- Now let's get into some programs implementing the DFT

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Defining the sequence of the samples of the signal
samples = np.array([1, 2, 3, 4])
N = len(samples)

#An array to store the results of the DFT
result = np.zeros(N, dtype=complex)
M = len(result)

#An array to store the various amplitudes
amp = np.zeros(N)
A = len(amp)

fs = 100
```

```python
#Implementing the formula: DFT
for k in range(N):
    for i in range(N):
        result[k] = result[k] + (samples[i] * np.exp((-1j * 2 * np.pi * k * i) / N))
```

```python
#Printing the values of the DFT
for n in range(M):
    print(f"{result[n]:.1f}")

#Calculating the values for the amplitude spectrum
f = np.zeros(N)
for a in range(A):
    amp[a] = 1/N * np.sqrt(((result[a].real)**2) + ((result[a].imag)**2))
    f[a] = (a * fs) / N #this calculated the corresponding frequencies

#Plotting the amplitude spectrum
plt.stem(f, amp)
plt.xlabel("f / Hz")
plt.ylabel("|X[k]|")
plt.grid(True)
```

- In this section, we only created our samples for the DFT. Next, it would be subjected it to a well defined signal: a pure sine wave.

- The formula to calculate the amplitude spectrum is:

$$A[k] = \frac{1}{N} \sqrt{\text{Re}(X[k])^2 + \text{Im}(X[k])^2}$$
